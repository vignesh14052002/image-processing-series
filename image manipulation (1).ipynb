```python
#adjust brightness

img1=np.copy(img)
x,y,z=img.shape
def bright(img1,b):
    if max(img1[i][j])+b>=255:
        img1[i][j]=255
    else:
        img1[i][j]+=b
def dull(img1,d):
    if min(img1[i][j])-d<=0:
        img1[i][j]=0
    else:
        img1[i][j]-=d

for i in range(x):
    for j in range(y):
        #place either dull or bright function here
        bright(img1,100)
opimg=Image.fromarray(img1)
opimg.show()
#author-@_._hello_world._._

```

```python
#flip images

inimg=Image.open(r'C:\Users\Home\Pictures\curvedarrow.jpg')
img=np.asarray(inimg)
x,y,z=img.shape
img1=img.copy()
def flip_vertical():
    global img,img1,x
    for i in range(x):
        img1[i,:]=img[x-i-1,:]
def flip_horizontal():
    global img,img1,y
    for i in range(y):
        img1[:,i]=img[:,y-i-1]
flip_vertical()
opimg=Image.fromarray(img1)
opimg.show()
#author-@_._hello_world._._
```

```python


inimg=Image.open(r'C:\Users\Home\Pictures\curvedarrow.jpg')
img=np.asarray(inimg)
x,y,z=img.shape
img1=img.copy()
def flip_vertical():
    global img,img1,x
    for i in range(x):
        img1[i,:]=img[x-i-1,:]
def flip_horizontal():
    global img,img1,y
    for i in range(y):
        img1[:,i]=img[:,y-i-1]
def rotate90():
    global img,img1,y
    img1=np.zeros((y,x,z),dtype='uint8')
    for i in range(y):
        img1[i,:]=img[:,i] 
def rotate180():
    global img,img1,x,y
    flip_vertical()
    img=img1.copy()
    flip_horizontal()
def rotate270():
    global img,img1,x,y,z
    rotate90()
    img=img1.copy()
    x,y=y,x
    flip_horizontal()
    img=img1.copy()
    flip_vertical()
    
rotate180()
opimg=Image.fromarray(img1)
opimg.show()   
```

```python
#box blur
from PIL import Image
import numpy as np
inimg=Image.open(r'C:\Users\Home\Pictures\gvr.jpg')
img=np.asarray(inimg)
x,y,z=img.shape
img1=np.zeros((img.shape),dtype='uint8')
def box_blur(n):
    global img,img1
    for i in range(0,x-n,n):
        for j in range(0,y-n,n):
            img1[i:i+n,j:j+n]+=img[i][j]
box_blur(10)
opimg=Image.fromarray(img1)
opimg.show()
#author-@_._hello_world._._
```

```python
#smooth blur
from PIL import Image
import numpy as np
inimg=Image.open(r'C:\Users\Home\Pictures\gvr.jpg')
img=np.asarray(inimg)
x,y,z=img.shape
img1=img.copy()
smooth =  [[1,1,1],
         [1,1,1],
         [1,1,1]]

def blur(kernel):
    global img,img1,x,y,z
    for i in range(x-3):
        for j in range(y-3):
            for k in range(3):
                img1[i][j][k]=(np.sum(img[i:i+3,j:j+3,k]*kernel))//np.sum(kernel)
    img=img1.copy()
for i in range(1):
    blur(smooth)
opimg=Image.fromarray(img1)
opimg.show()
#author-@_._hello_world._._

```

```python
#gaussian kernel
import numpy as np
n=7 #use only odd number
g=np.ones((n,n),dtype='int')

for i in range(1,n):
    #row
    if i<=n//2:
        g[0][i]=g[0][i-1]*2
    else:
        g[0][i]=g[0][i-1]//2
    #coloumn
    if i<=n//2:
        g[i][0]=g[i-1][0]*2
    else:
        g[i][0]=g[i-1][0]//2
    
for i in range(1,n):    
    for j in range(1,n):
        g[i][j]=g[0][j]*g[i][0]
        
print(g)
```
